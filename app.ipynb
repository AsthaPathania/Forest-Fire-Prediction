````python
dataset.loc[:124,"region"]=0
dataset.loc[125:,"region"]=1
df=dataset
````

````python
df.info()
````

dataset – the DataFrame you are working with.

.loc[] – a label-based indexer used to select rows and columns. Its general form is:

df.loc[rows, columns]


rows – what rows to select.

:124 means all rows from 0 to 124 inclusive.

columns – which column(s) to select or create.

"region" here is the column label.

If "region" does not exist, Pandas automatically creates it.

= 0 – the value to assign to the selected rows in that column.

Key idea

In Pandas, if you try to assign values to a non-existing column, it creates a new column with that name.

So "region" was not originally in the dataset, but now it exists because you assigned values to it.
